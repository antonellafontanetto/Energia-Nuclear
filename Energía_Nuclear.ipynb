```python
import folium
from folium.plugins import MarkerCluster
from branca.element import Template, MacroElement

# Crear mapa base
mapa = folium.Map(location=[20, 0], zoom_start=2, tiles='CartoDB dark_matter')

# Agrupar marcadores
marker_cluster = MarkerCluster().add_to(mapa)

# Agregar cada reactor
for _, row in df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    name = row['Plant']

    if pd.notna(lat) and pd.notna(lon):
        folium.Marker(
            location=[lat, lon],
            popup=f"<b>{name}</b>",
            tooltip=name,
            icon=folium.Icon(color='green', icon='info-sign')
        ).add_to(marker_cluster)

# Agregar t√≠tulo al mapa (HTML embebido)
titulo_html = """
{% macro html(this, kwargs) %}
<div style="position: fixed;
            top: 20px; left: 50px; width: 70%;
            z-index:9999;
            font-size:24px;
            font-weight: bold;
            color: white;
            background-color: rgba(0,0,0,0.3);
            padding: 10px;
            border-radius: 8px;">
    üåç Mapa Mundial de Reactores Nucleares
</div>
{% endmacro %}
"""

template = Template(titulo_html)
macro = MacroElement()
macro._template = template
mapa.get_root().add_child(macro)

# Mostrar el mapa
mapa
```

```python
import folium
from folium.plugins import MarkerCluster
from branca.element import Template, MacroElement

# Crear mapa base con tama√±o reducido
mapa = folium.Map(
    location=[20, 0],
    zoom_start=2,
    tiles='CartoDB dark_matter',
    width='70%',       # Ancho reducido
    height='450px'     # Altura reducida
)

# Agrupar marcadores
marker_cluster = MarkerCluster().add_to(mapa)

# Agregar cada reactor
for _, row in df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    name = row['Plant']

    if pd.notna(lat) and pd.notna(lon):
        folium.Marker(
            location=[lat, lon],
            popup=f"<b>{name}</b>",
            tooltip=name,
            icon=folium.Icon(color='green', icon='info-sign')
        ).add_to(marker_cluster)

# Agregar t√≠tulo HTML al mapa
titulo_html = """
{% macro html(this, kwargs) %}
<div style="position: fixed;
            top: 20px; left: 50px; width: auto;
            z-index:9999;
            font-size:20px;
            font-weight: bold;
            color: white;
            background-color: rgba(0,0,0,0.4);
            padding: 10px;
            border-radius: 8px;">
    üåç Mapa Mundial de Reactores Nucleares
</div>
{% endmacro %}
"""

template = Template(titulo_html)
macro = MacroElement()
macro._template = template
mapa.get_root().add_child(macro)

# Mostrar el mapa
mapa
```
